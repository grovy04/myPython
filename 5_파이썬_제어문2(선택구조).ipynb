### if...
- 조건이 참일 때에는 아래 명령문을 실행하고, 거짓일 때에는 아무것도 하지 않는 가장 단순한 형태의 조건문
```
  if 조건 :
    조건이 참일 때 실행할 명령어1
    조건이 참일 때 실행할 명령어2
```

````python
a = int(input('수를 입력하세요:'))

if a > 0:
  print('양수입니다')

````

````python
a = int(input('1 + 2 는? '))

if a == 3:
  print('정답입니다')
else:
  print('오답입니다')
````

### * <font size=5 color='tomato'>확인문제 6 </font>
짝수와 홀수를 구분할 수 있는 프로그램을 만들어보자

````python
a = int(input('수를 입력하세요:'))
if a % 2 == 0:
  print('짝수입니다.')
if a % 2 == 1:
  print('홀수입니다.')

````

### if..else..
- 조건이 참일 때와 거짓일 때 실행할 명령어 코드가 존재하는 경우 사용
```
  if 조건 :
    조건이 참일 때 실행할 명령어1   
  else :
    조건이 거짓일 때 실행할 명령어1
```

````python
a = int(input('수를 입력하세요:'))

if a % 2 == 0:
  print('짝수')
else:
  print('홀수')
````

````python
a = int(input('수를 입력하세요:'))

if a == 0:
  print('0')
elif a < 0:
  print('음수')
else:
  print('양수')



````

### * <font size=5 color='tomato'>확인문제 7 </font>
성적을 입력 받는다

  - 90점 이상이면 "A"
  - 80점 이상이면 "B"
  - 70점 이상이면 "C"
  - 아니면 "D"

````python
score = int(input('성적을 입력하세요 : '))
if score >= 90:
  print('A')
else:
  if score >= 80:
    print('B')
  else:
    if score >= 70:
      print('C')
    else:
      print('D')

````

### if...elif...else...
- 중첩if
- if문 안에 또 다른 if문이 있는 형태

````python
score = int(input('성적을 입력하세요 : '))
if score >= 90:
  print('A')
elif score >= 80:
  print('B')
elif score >= 70:
  print('C')
else:
  print('D')
````

### * <font size=5 color='tomato'>확인문제 8 </font>
- 나이를 입력받아 국립공원 입장권을 발급하는 프로그램
  - 기본 입장료는 3000원
  - 나이가 65세 이상이거나 7세 이하인 경우에는 무료
  - 나이 8세~18세는 청소년 요금인 1000원
  - 나이에 따라 입장료를 출력하세요


image.png

````python
#확인문제
age = int(input('나이를 입력하세요 : '))

fee = 3000

if 8 <= age <= 18:
  fee == 1000
elif age >= 65 or age <= 7:
  fee = 0

print(f'입장료는 {fee}입니다')
````
