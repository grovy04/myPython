#### 문자열.strip(지울 문자열)
  - 문자열의 선행(앞),후행(뒤)의 공백 문자를 제거한 후 반환

### 불린(bool)
- 주어진 조건을 만족하는지 여부를 참(True)과 거짓(False)으로 표현하기 위한 자료형

````python
#수식 이용
print(5 > 2)
print(5 == 2)
print(5 == str(5))
````

### match-case문
- python 3.10이상부터 다른 언어의switch-case문과 비슷한 기능을 제공하는 match-case문을 사용할 수 있다

````python
for i in range(1, 11):
    match i % 2:
        case 0:
            print(f'{i}는 짝수입니다.')
        case 1:
            print(f'{i}는 홀수입니다.')
````
