### while반복
-조건이 만족하는 동안 반복

image.png

```python
#True인 동안 반복된다(False가 되면 반복에서 벗어남)
while True:
  print('never ending story')
```

```python
#순차 : print문과 count++문의 순서를 바꿔가면 실행해보기
boolValue = True
count = 0

while boolValue:
  print(count)
  count += 1
  if count == 10:
    boolValue = False
```

```python
#반복 변수를 기반으로 반복
i = 0
while i < 10:
  print(f'{i}번째 반복입니다')
  i += 1
```

```python
family = ['할아버지','할머니','누나','동생']
i = 0
while i < len(family):
  print('우리 가족에는',family[i],'가(이) 있습니다.')
  i = i+1
```

#### <font size=5 color='tomato'>확인문제 6</font>    
- 성적을 계속 입력 받아 scores라는 리스트 변수에 추가
- -1을 입력하면 종료
- 입력이 종료되면 scores리스트에 있는 성적들의 합계와 평균을 구하는 프로그램을 작성

```python
scores = []
s = 0
sum = 0

while s != -1:
  s = int(input('성적을 입력하세요(-1을 입력하면 종료) : '))
  if s == -1 :
    break
  sum += s
  scores.append(s)
print(f'합계 : {sum}, 평균 : {sum/len(scores)}')
```

### 반복문을 벗어나기 위한 문장

####break
- break: 조건문을 벗어남

```python
sum = 0

for i in range(1, 11, 1):
  print(f'{i}번 반복했습니다')
  sum += i
  if i == 5:
    break

print(f'총 합계는 {sum}입니다')
```

#### continie
- 다음 문장을 실행하지 않고, 반복문의 처음으로 돌아가서 계속 반복을 수행한다.

```python
sum = 0

for i in range(1, 11, 1):
  if i == 3:
    continue
  print(f'{i}번 반복했습니다')
  sum += i


print(f'총 합계는 {sum}입니다')
```
