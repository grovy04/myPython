## 리스트(List)
- 여러가지 자료를 저장할 수 있는 목록으로 []안에 자료를 넣으며, 내부 요소는 `,`로 구분한다
- 리스트 안에 들어가는 여러가지 자료를 요소(element)라고 한다
- 각각의 요소는 유의미한 순서를 갖는다
  
  0부터 순서대로 번호를 붙여 사용하고 이것을 인덱스(index)라고 부른다 (`리스트[인덱스]`)
- 숫자, 문자열, 불린의 기본 자료형 여러 개를 하나의 리스트에 저장할 수 있다

#### 생성 / 수정 / 삭제
    리스트 생성 방법

    1) 리스트이름 = [요소1, 요소2, 요소3]
    2) 리스트이름 = []
      리스트이름.append('요소1')
      리스트이름.append('요소2')
      리스트이름.append('요소3')
    3) 리스트이름 = list(range(1,10,1))  
    4) 리스트이름 = list()
#### 수정
    1) 리스트이름[인덱스]=수정된 값
#### 삭제
    1) 리스트이름.remove('삭제할 요소')   
    2) del(리스트이름[인덱스])  
    3) 리스트이름.pop()  

````python
num = 0
hap = 0
for i in range(0,4):
  num = int(input(f'{i+1}번째 숫자를 입력하세요 : '))
  hap += num
print(hap)
````

````python
#예제 1 --- 이것은 같이 만들어 볼 예정임
numList = []
hap = 0
for i in range(0,4):
  numList.append(int(input(f'{i+1}번째 숫자를 입력하세요 : ')))
  hap += numList[i]
print(hap)
````

#### 얕은 복사(shallow copy)와 깊은 복사(deep copy)
  - 리스트의 b에 a를 할당하면 값이 할당되는 것이 아니라 같은 메모리 주소를 바라보게 된다.
  - a(또는 b)를 변경하면 b(또는 a)도 같이 바뀐다
  - mutable한 다른 객체 또한 똑같은 현상이 나타난다
  - 이런 현상은 deep copy를 사용하여 해결한다

image.png

````python
#shallow copy
a = [2,5]
b = a
a.append(7)
print(a,b)
````

````python
#deep copy
d = a.copy()
a.append(8)
print(a,d)
````

#### <font size=5 color='tomato'>확인문제 2</font>
```
singer = ['멜로망스', '잔나비', '이센스', '10센티', '이영지']
  1) 맨 뒤에 '비오'를 추가하세요
  2) '잔나비'뒤에 '딥플로우'를 추가하세요
  3) '지코', '저스디스'를 한 번에 추가하세요
  4) '지코'를 제거하세요
  5) 리스트를 모두 제거하세요
```


````python
singer = ['멜로망스', '잔나비', '이센스', '10센티', '이영지']

singer.append('비오') #맨 뒤에 '비오'를 추가하세요
print(singer)
singer.insert(2, '딥플로우') #'잔나비'뒤에 '딥플로우'를 추가하세요
print(singer)
singer.extend(['지코', '저스디스']) #'지코', '저스디스'를 한 번에 추가하세요
print(singer)
singer.remove('지코') #'지코'를 제거하세요
print(singer)
singer.clear() #리스트를 모두 제거하세요
print(singer)
````
