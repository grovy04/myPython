###연산자의 우선순위
  -  1. `(), [], {}` : 괄호, 리스트, 딕셔너리, 세트 등

  -  5. `+, -` : 산술 연산자
  -  6. <<, >> : 비트 시프트 연산자

  -  8. ^ : 비트 배타적 논리합
  -  9. | : 비트 논리합
  - 10. `<, >, >=, <=` : 관계 연산자
  - 11. `==, !=` : 동등 연산자
  - 12. ` =, %=, /=, //=, -=, +=, *=, **=` : 대입 연산자
  - 13. `not` : 논리연산자
  - 14. `and` : 논리연산자
  - 15. `or` : 논리연산자

#### 문자열.strip(지울 문자열)
  - 문자열의 선행(앞),후행(뒤)의 공백 문자를 제거한 후 반환

### 불린(bool)
- 주어진 조건을 만족하는지 여부를 참(True)과 거짓(False)으로 표현하기 위한 자료형

비교연산자 : 값의 크기를 비교하는 연산자로 조건에 만족하면 참(True)을 만족하지 않으면 거짓(False)을 반환

````python
#불린형 변수 이용
a = (5 > 2)
b = (5 < 2)
print(a, type(a))
print(b, type(b))
````

````python
#수식 이용
print(5 > 2)
print(5 == 2)
print(5 == str(5))
````

논리연산자 : and, or, not으로 연산결과에 따라 참(True)과 거짓(False)을 반환한다

````python
print( 5 > 2 and 7 < 4)
print( 5 > 2 or 7 < 4)
print(not 5 > 2)
````
